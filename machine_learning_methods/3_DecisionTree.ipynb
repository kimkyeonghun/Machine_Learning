```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = './data/IRIS.csv'

raw = pd.read_csv(file)
raw.head()
```

data를 서로 구분할 수 있도록 unique한 컬럼인 id를 추가한다.

```python
raw['species'].unique()
raw['id'] = range(len(raw))
raw
```

## train_test_split

```python
train = raw.sample(100, replace=False, random_state=42)
```

```python
train = train.reset_index(drop=True)
train
```

```python
#train에 포함되지 않은 값들을 뽑는다.
test = raw.loc[~raw['id'].isin(train['id'])]
test = test.reset_index(drop=True)
test
```

## 2.의사결정나무 실습

**특징**
1. 해석이 쉽다.
2. 빠르다.

**파라미터**
1. min_samples_split: 의사결정나무에서 각 구간의 최소 샘플 수, 이보다 작으면 더 이상 분리되지 않음

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(min_samples_split = 10)
```

```python
input_col = ['sepal_length','sepal_width','petal_length','petal_width']
dt.fit(train[input_col],train['species'])
```

```python
predicts = dt.predict(test[input_col])
test['pred'] = predicts
test
```

```python
(test['pred']==test['species']).mean()
```
